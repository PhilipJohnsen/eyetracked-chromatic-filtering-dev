# Foveal settings value calculation


This file acts as a calculator for the, maximum values for foveal radius and transition width as float[0,1] that still satisfies the estimated foveal radius in which the medical effect should still be effective. 

The float[0,1] is a percentage of the screens width to use for masking out the gaussian blur.

```python
import math

#Screen physical parameters
screen_size_diag_mm  = 686.0    #27 inches = 686mm diagonal
aspect_ratio         = 16 / 9   #W/H ratio 
dist_to_screen_mm    = 500.0    #Eye to screen distance

#Foveal region (degrees of visual angle)
foveal_degree_radius    = 6.0   #Sharp foveal zone
transition_degree_width = 6.0   #Blur-transition ring
```

### Calculation

Converting **visual angle to physical size on screen to NDC**:

**1. Screen width**
$a = W/H$:
$$W_\text{screen} = d_\text{diag} \cdot \frac{a}{\sqrt{a^2 + 1}}$$

**2. Physical radius from visual angle**
The length $r$ subtended by angle $\theta$ at viewing distance $d$:
$$r = d \cdot \tan(\theta)$$

**3. NDC**
The shader uses NDC $[0, 1]$

foveal = $\frac{r_\text{foveal}}{W_\text{screen}}$

transition = $\frac{r_\text{transition}}{W_\text{screen}}$

```python
#Screen width / height in mm
screen_width_mm  = screen_size_diag_mm * aspect_ratio / math.sqrt(aspect_ratio**2 + 1)
screen_height_mm = screen_width_mm / aspect_ratio

#Physical radii on screen (mm)
foveal_radius_mm    = dist_to_screen_mm * math.tan(math.radians(foveal_degree_radius))
transition_width_mm = dist_to_screen_mm * math.tan(math.radians(transition_degree_width))

#NDC, fraction of screen width
foveal_radius    = foveal_radius_mm    / screen_width_mm  
transition_width = transition_width_mm / screen_width_mm 
```

### Results

The values below are the values needed for /render/utility/settings.txt


```python
print(f"Screen size       : {screen_width_mm:.1f} × {screen_height_mm:.1f} mm")
print(f"Viewing distance  : {dist_to_screen_mm:.0f} mm")
print()
print(f"Foveal radius     : {foveal_degree_radius}° → {foveal_radius_mm:.2f} mm → foveal_radius    = {foveal_radius:.4f}")
print(f"Transition width  : {transition_degree_width}° → {transition_width_mm:.2f} mm → transition_width = {transition_width:.4f}")
```
